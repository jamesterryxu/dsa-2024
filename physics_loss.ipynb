```python
import numpy as np
import random
x_position = np.linspace(0, 100, 5)
y_deflection = random.sample(range(1, 10), 5)
```

```python
# x_position: the sequence of prsitions
# y_deflection: the sequence of deflections
# k: stifness EI
# force=force
def phy_loss_fun(x_position,y_deflection,k,force):
    x_len=x_position[1]-x_position[0]
    derivate=np.gradient(y_deflection,x_len, axis=None, edge_order=2)
    phy_loss=sum(derivate*k-force*(x_position[-1]-x_position))
    return phy_loss
```

```python
a=phy_loss_fun(x_position,y_deflection,8,1)
print(a)
```
